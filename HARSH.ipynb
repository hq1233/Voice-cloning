```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio

preload_models()
```

```python
text_prompt = """
     Hello, my name is Harsh, currently pursuing my bachelor’s of technology in computer science and engineering from SRM institute of science and technology. i am a web developer and ai/ml enthusiast.
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```
